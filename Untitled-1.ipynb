```python
import asyncio
from twscrape import API, gather
from twscrape.logger import set_log_level
import json
import time
from langdetect import detect, LangDetectException
import re


search_term = "Tesla OR TSLA"
async def main():
    api = API()  # or API("path-to.db") - default is `accounts.db`

    # ADD ACCOUNTS (for CLI usage see BELOW)
    await api.pool.add_account("DarOho4050", "asdfghjkl1!", "we444465@gmail.com", "mail_pass1")
    # await api.pool.add_account("user2", "pass2", "u2@example.com", "mail_pass2")
    await api.pool.login_all()

# search (latest tab)
    tweets = await gather(api.search(search_term, limit=50))  # list[Tweet]
    print(tweets)
    
    def user_to_dict(user):
        return {
            'id': user.id,
            'id_str': user.id_str,
            'url': user.url,
            'username': user.username,
            'displayname': user.displayname,
            'rawDescription': user.rawDescription,
            'created': user.created.isoformat() if user.created else None,            
        }
    
    def tweet_to_dict(tweet):
        return {
            'id': tweet.id,
            'date': tweet.date.isoformat(),
            'rawContent': tweet.rawContent,
            'url': tweet.url,
            'user': user_to_dict(tweet.user),
            'lang': tweet.lang,
            'replyCount': tweet.replyCount,
            'retweetCount': tweet.retweetCount,
            'likeCount': tweet.likeCount,
            'quoteCount': tweet.quoteCount
        }

    # Convert each Tweet instance to a dictionary
    tweets_as_dicts = [tweet_to_dict(tweet) for tweet in tweets]
    print(tweets_as_dicts)
    return tweets_as_dicts
    
def write_tweets_to_file(tweets, file_path):
    """Generic function to write tweets to a specified file with deduplication."""
    new_tweets_count = 0
    existing_ids = set()
    try:
        with open(file_path, 'r') as file:
            existing_ids.update(json.loads(line)['id'] for line in file)
    except FileNotFoundError:
        pass  # It's okay if the file does not exist yet

    with open(file_path, 'a') as file:
        for tweet in tweets:
            if tweet['id'] not in existing_ids:
                file.write(json.dumps(tweet) + '\n')
                new_tweets_count += 1
                existing_ids.add(tweet['id'])

    return new_tweets_count
    
def clean_tweet_content(text):
    """Remove mentions and links from tweet text."""
    text = re.sub(r'@\w+', '', text)  # Remove mentions
    text = re.sub(r'https?://\S+', '', text)  # Remove URLs
    return text.strip()

def matches_search_criteria(tweet_content, search_terms):
    """Check if tweet content matches the search criteria using regex for flexibility."""
    # For simplicity, this example focuses on 'OR' logic. It can be extended for 'AND'.
    search_terms = search_terms.replace(' OR ', '|')  # Replace 'OR' with regex '|' for alternation
    pattern = r'\b(' + search_terms + r')\b'  # Word boundary regex to match whole words only
    return re.search(pattern, tweet_content, re.IGNORECASE) is not None

def filter_tweets(tweets):
    english_tweets = []
    filtered_tweets = []  # Tweets filtered out
    for tweet in tweets:
        # Remove mentions from the tweet's rawContent
        cleaned_content = clean_tweet_content(tweet['rawContent'])
        if matches_search_criteria(cleaned_content.lower(), search_term.lower()):
            try:
                if detect(cleaned_content) == 'en':
                    english_tweets.append(tweet)
                else:
                    filtered_tweets.append(tweet)
            except LangDetectException:
                # Assume English if detection fails
                english_tweets.append(tweet)
    return english_tweets, filtered_tweets

def write_english_tweets(tweets):
    """Write English tweets to their file."""
    return write_tweets_to_file(tweets, 'english_tweets.jsonl')

def write_filtered_tweets(tweets):
    """Write filtered (non-English) tweets to their file."""
    return write_tweets_to_file(tweets, 'filtered_tweets.jsonl')


while True:
    tweets = await main()  # Assume this fetches a list of tweet dictionaries
    english_tweets, filtered_tweets = filter_tweets(tweets)
    
    new_english_count = write_english_tweets(english_tweets)
    filtered_count = write_filtered_tweets(filtered_tweets)
    
    total_processed = len(english_tweets) + len(filtered_tweets)
    
    if total_processed > 0:
        english_percentage = (new_english_count / total_processed) * 100
        filtered_percentage = (filtered_count / total_processed) * 100
        print(f"Processed {total_processed} tweets. New English: {english_percentage:.2f}%. Filtered: {filtered_percentage:.2f}%.")
    else:
        print("No tweets to process.")

    time.sleep(90)  # Sleep for x seconds
    
if __name__ == "__main__":
    tweets = await main()  # Should return a list of tweet dictionaries
new_tweets_count, total_tweets = write_tweets_to_jsonl(tweets)
    
if total_tweets > 0:  # Prevent division by zero
    new_tweets_percentage = (new_tweets_count / total_tweets) * 100
    print(f"New tweets: {new_tweets_count} of {total_tweets} - {new_tweets_percentage:.2f}% new")
else:
    print("No tweets to process.")
```
